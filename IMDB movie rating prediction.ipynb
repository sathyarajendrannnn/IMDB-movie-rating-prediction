**DATA PREPROCESSING**

```python
print(mf['Year'].dtype)  
mf['Year'] = mf['Year'].astype(str)  
mf['Year'] = mf['Year'].str.replace(r'[^\d]', '', regex=True).astype(int)  
```

```python
mf['Duration'] = mf['Duration'].astype(str).str.replace('min', '', regex=False)
```

```python
mf['Genre'] = mf['Genre'].str.split(', ')
mf = mf.explode('Genre')
mf['Genre'].fillna(mf['Genre'].mode()[0], inplace=True)
```

```python
mf['Votes'] = mf['Votes'].str.replace(r'[\$,]', '', regex=True)

mf['Votes'] = pd.to_numeric(mf['Votes'], errors='coerce')
```

```python
avg_rating_by_year= mf.groupby(['Year','Genre'])['Rating'].mean().reset_index()
top_genres =mf['Genre'].value_counts().head(10).index
average_rating_by_year = avg_rating_by_year[avg_rating_by_year['Genre'].isin(top_genres)]
fig=px.line(average_rating_by_year,x='Year',y='Rating',color="Genre")
fig.update_layout(title='Average Rating by Year for Top Genres', xaxis_title='Year', yaxis_title='Average Rating')
fig.show()
```
